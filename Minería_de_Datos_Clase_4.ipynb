```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Crear DataFrame
df = pd.DataFrame({
    "Cliente": [1, 2, 3, 4, 5, 6, 7, 8],
    "Edad": [25, 40, 35, 50, 30, 45, 38, 29],
    "Ingresos": [50000, 60000, 70000, 80000, 45000, 85000, 72000, 48000],
    "Compras": [5, 15, 10, 20, 7, 18, 12, 6],
    "Categoría": ["A", "B", "C", "A", "B", "C", "A", "B"]
})

df
```

```python
# Histograma de Edad
sns.histplot(df["Edad"], bins=5, kde=True)
plt.title("Distribución de Edades")
plt.show()

# Histograma de Ingresos
sns.histplot(df["Ingresos"], bins=5, kde=True)
plt.title("Distribución de Ingresos")
plt.show()
```

```python
# Boxplot de Compras
sns.boxplot(x=df["Compras"])
plt.title("Boxplot de Compras")
plt.show()
```

```python
# Scatterplot entre Ingresos y Compras
sns.scatterplot(x=df["Ingresos"], y=df["Compras"])
plt.title("Relación entre Ingresos y Compras")
plt.xlabel("Ingresos")
plt.ylabel("Compras")
plt.show()
```

```python
# Pairplot
sns.pairplot(df[["Edad", "Ingresos", "Compras"]])
plt.show()
```

```python
# Heatmap de correlaciones
correlaciones = df.corr(numeric_only=True)

sns.heatmap(correlaciones, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Mapa de Correlación de Variables")
plt.show()
```

```python
# Violin Plot de Compras por Categoría
sns.violinplot(x="Categoría", y="Compras", data=df)
plt.title("Distribución de Compras por Categoría")
plt.show()
```

# Conclusión

Primero se realizaron histogramas para observar la distribución de Edad e Ingresos. Después se utilizó un boxplot para analizar la variable Compras y detectar posibles valores atípicos. También se hizo un scatterplot para ver la relación entre Ingresos y Compras.

Luego se aplicaron visualizaciones avanzadas como pairplot, heatmap y violin plot. Estas herramientas permitieron observar relaciones entre las variables y comparar las compras según la categoría.
