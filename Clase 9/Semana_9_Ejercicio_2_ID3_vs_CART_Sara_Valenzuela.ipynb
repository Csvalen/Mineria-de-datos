## Objetivo

Comparar el desempeño de los algoritmos ID3 y CART utilizando el dataset de pingüinos (Penguins) para clasificar la especie de cada ejemplar.

```python
# Importamos Librerias
import pandas as pd
import seaborn as sns
import numpy as np

from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
```

```python
# Cargar dataset de pingüinos
df = sns.load_dataset("penguins")

# Eliminar valores faltantes
df = df.dropna()

# Visualizar primeras filas
df.head()
```

```python
# Variables predictoras
X = df[[
    "bill_length_mm",
    "bill_depth_mm",
    "flipper_length_mm",
    "body_mass_g"
]]

# Variable objetivo
y = df["species"]

# División entrenamiento y prueba
X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    test_size=0.20,
    random_state=42
)

print("Entrenamiento:", X_train.shape)
print("Prueba:", X_test.shape)
```

## Preparación de los datos

Se utilizaron variables físicas de los pingüinos como longitud y profundidad del pico, longitud de la aleta y masa corporal para predecir la especie de cada ejemplar.

El conjunto de datos fue dividido en entrenamiento (80%) y prueba (20%) para evaluar la capacidad de generalización de los modelos.

```python
# Árbol ID3 (Entropía)

tree_id3 = DecisionTreeClassifier(
    criterion="entropy",
    max_depth=4,
    random_state=42
)

tree_id3.fit(X_train, y_train)

y_pred_id3 = tree_id3.predict(X_test)

acc_id3 = accuracy_score(y_test, y_pred_id3)

print("Precisión ID3:", acc_id3)
```

```python
# Árbol CART (Gini)

tree_cart = DecisionTreeClassifier(
    criterion="gini",
    max_depth=4,
    random_state=42
)

tree_cart.fit(X_train, y_train)

y_pred_cart = tree_cart.predict(X_test)

acc_cart = accuracy_score(y_test, y_pred_cart)

print("Precisión CART:", acc_cart)
```

```python
comparacion = pd.DataFrame({
    "Modelo": ["ID3", "CART"],
    "Accuracy": [acc_id3, acc_cart]
})

comparacion
```

## Conclusión Final

Se compararon dos algoritmos de árboles de decisión utilizando el dataset de pingüinos.

El modelo ID3, basado en Entropía, obtuvo una precisión de 98,51%, mientras que el modelo CART, basado en el Índice Gini, alcanzó exactamente el mismo resultado.

Esto indica que ambos criterios de división fueron igualmente efectivos para clasificar las especies presentes en este conjunto de datos. La elevada precisión obtenida demuestra que las variables seleccionadas poseen una fuerte capacidad discriminatoria para diferenciar las distintas especies de pingüinos.

Aunque los criterios matemáticos utilizados por ID3 y CART son diferentes, en este caso ambos algoritmos produjeron un desempeño prácticamente idéntico. Esto evidencia que la elección entre Entropía y Gini no siempre genera diferencias significativas cuando los datos presentan una buena separación entre clases.
