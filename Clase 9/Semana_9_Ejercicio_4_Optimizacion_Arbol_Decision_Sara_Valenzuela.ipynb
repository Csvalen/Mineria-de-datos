## Objetivo

Optimizar un árbol de decisión utilizando GridSearchCV para encontrar automáticamente la mejor combinación de hiperparámetros.

```python
# Importamos Librerias

import pandas as pd
import seaborn as sns

from sklearn.model_selection import (
    train_test_split,
    GridSearchCV
)

from sklearn.tree import DecisionTreeClassifier

from sklearn.metrics import (
    accuracy_score,
    classification_report
)
```

```python
# Cargar dataset
df = sns.load_dataset("penguins")

# Eliminar valores faltantes
df = df.dropna()

df.head()
```

```python
# Variables predictoras
X = df[[
    "bill_length_mm",
    "bill_depth_mm",
    "flipper_length_mm",
    "body_mass_g"
]]

# Variable objetivo
y = df["species"]

# División entrenamiento y prueba
X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    test_size=0.20,
    random_state=42
)

print("Entrenamiento:", X_train.shape)
print("Prueba:", X_test.shape)
```

```python
# Variables predictoras
X = df[[
    "bill_length_mm",
    "bill_depth_mm",
    "flipper_length_mm",
    "body_mass_g"
]]

# Variable objetivo
y = df["species"]

# División entrenamiento y prueba
X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    test_size=0.20,
    random_state=42
)

print("Entrenamiento:", X_train.shape)
print("Prueba:", X_test.shape)
```

## Preparación de los datos

Se utilizó el dataset Penguins para construir un árbol de decisión orientado a clasificar especies de pingüinos.

Posteriormente se aplicó GridSearchCV para identificar automáticamente la combinación de parámetros que produce el mejor rendimiento.

```python
modelo_base = DecisionTreeClassifier(
    random_state=42
)

modelo_base.fit(X_train, y_train)

pred_base = modelo_base.predict(X_test)

acc_base = accuracy_score(
    y_test,
    pred_base
)

print("Accuracy modelo base:", acc_base)
```

```python
param_grid = {
    "max_depth": [2, 3, 4, 5, 6],
    "criterion": ["gini", "entropy"],
    "min_samples_split": [2, 5, 10]
}

grid = GridSearchCV(
    DecisionTreeClassifier(random_state=42),
    param_grid,
    cv=5,
    scoring="accuracy"
)

grid.fit(X_train, y_train)

print("Mejores parámetros:")
print(grid.best_params_)
```

```python
mejor_modelo = grid.best_estimator_

pred_opt = mejor_modelo.predict(X_test)

acc_opt = accuracy_score(
    y_test,
    pred_opt
)

print("Accuracy optimizado:", acc_opt)
```

```python
comparacion = pd.DataFrame({
    "Modelo": ["Base", "Optimizado"],
    "Accuracy": [acc_base, acc_opt]
})

comparacion
```

```python
print(classification_report(
    y_test,
    pred_opt
))
```

## Conclusión Final

Se aplicó GridSearchCV para optimizar automáticamente los hiperparámetros de un árbol de decisión mediante validación cruzada.

El proceso identificó como mejor configuración un árbol con criterio Gini, profundidad máxima de 6 niveles y un mínimo de 2 observaciones para realizar divisiones internas.

Tanto el modelo base como el modelo optimizado alcanzaron una precisión del 100% sobre el conjunto de prueba. Esto indica que el dataset presenta una separación muy clara entre las distintas especies de pingüinos, permitiendo que el árbol de decisión realice clasificaciones perfectas.

Los resultados obtenidos demuestran la utilidad de GridSearchCV para automatizar la búsqueda de hiperparámetros óptimos y validar el rendimiento de los modelos de clasificación.
